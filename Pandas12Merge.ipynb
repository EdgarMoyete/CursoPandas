### pandas merge

```python
import pandas as pd
```

```python
df1 = pd.DataFrame([
    ["12", 3],
    ["13", 5],
    ["15", 8],
    ["16", 20],
    ["20", 10],
],
    columns=["id", "stock"])
df1
```

```python
df2 = pd.DataFrame([
    ["12", "100"],
    ["16", "200"],
    ["50", "300"]
],
    columns=["id", "precio"])
df2
```

```python
# por defecto how="inner" (filas comunes)
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how="outer")
```

### varias columnas

```python
df1 = pd.DataFrame([
    ["enero", "pan", 100],
    ["enero", "galletas", 50],
    ["febrero", "pan", 20],
    ["febrero", "galletas", 10]
],
    columns=["mes", "producto", "stock"])
df1
```

```python
df2 = pd.DataFrame([
    ["enero", "pan", 4],
    ["enero", "galletas", 8],
    ["febrero", "pan", 4.5],
    ["febrero", "galletas", 9]
],
    columns=["mes", "producto", "precio"])
df2
```

```python
pd.merge(df1, df2)
```

### columnas con nombres diferentes

```python
df1 = pd.DataFrame([
    ["pan", 3],
    ["galletas", 4],
    ["agua", 5],
    ["refresco", 6]
],
    columns=["producto", "stock"])
df1
```

```python
df2 = pd.DataFrame([
    ["pan", 13],
    ["galletas", 14],
    ["agua", 15]
],
    columns=["product", "precio"])
df2
```

```python
# no existen columnas con el mismo nombre
# pd.merge(df1, df2)
```

```python
df = pd.merge(df1, df2, left_on="producto", right_on="product")
df
```

```python
df.drop(["product"], axis=1)
```

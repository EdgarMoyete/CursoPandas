### Seleccionar y filtrar datos
#### por columnas, iloc, loc y condiciones

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

#### por columnas

```python
df["sepal length (cm)"]
```

```python
df[["sepal length (cm)"]]
```

```python
df[["sepal length (cm)", "petal width (cm)"]]
```

#### iloc indexando solo las filas

```python
df.iloc[2]
```

```python
df.iloc[[2]]
```

```python
df.iloc[[1,2]]
```

```python
df.iloc[:5]
```

```python
df.iloc[1:5]
```

#### iloc indexando ambos ejes

```python
df.iloc[3,2]
```

```python
df.iloc[:,2]
```

```python
df.iloc[1, :]
```

```python
df.iloc[0:11, 2]
```

```python
df.iloc[0:3, 1:3]
```

```python
df.iloc[[0,2,4,6,8],[1,2]]
```

#### loc

```python
df.loc[[2,3,4], ["sepal length (cm)", "sepal width (cm)"]]
```

#### condiciones

```python
df["petal width (cm)"] == 0.2
```

```python
df["petal length (cm)"] > 1.5
```

#### loc con condiciones

```python
df.loc[df["petal width (cm)"] == 0.2]
```

```python
df.loc[df["petal length (cm)"] > 1.5]
```

```python
df.loc[(df["petal width (cm)"] == 0.2) & (df["petal length (cm)"] > 1.5)]
```

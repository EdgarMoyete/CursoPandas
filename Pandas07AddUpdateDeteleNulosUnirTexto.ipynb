```python
import pandas as pd
```

## Agregar filas

```python
frutas = [["Platano", 3],
         [ "Kiwi", 4],
         [ "Manzana", 5],
         [ "Pera", 4]]
df_frutas = pd.DataFrame(frutas, columns=["Fruta", "Precio"])
df_frutas
```

```python
df_frutas = df_frutas.append({"Fruta" : "Limon", "Precio" : 1},
                             ignore_index=True)
df_frutas
```

```python
nuevas_frutas = [["Mango", 8],
                 [ "Naranja", 2]]
df_nuevo = pd.DataFrame(nuevas_frutas, columns=["Fruta", "Precio"])
df_nuevo
```

```python
df_frutas = df_frutas.append(df_nuevo, ignore_index=True)
df_frutas
```

```python
nuevas_frutas = [["Sandia", 12],
                 [ "Melon", 10]]
df_nuevo = pd.DataFrame(nuevas_frutas, columns=["Fruta", "Precio"])
df_nuevo
```

```python
df_frutas = pd.concat([df_frutas, df_nuevo], ignore_index=True)
df_frutas
```

## Agregar columnas

```python
df_frutas["Descuento"] = [0.5, 0.4, 0.3, 0.3, 0.4, 0.5, 0.1, 0.2, 0.3]
df_frutas
```

```python
df_frutas = df_frutas.assign(Stock = [2,2,4,5,6,1,2,3,4])
df_frutas
```

```python
df_frutas = df_frutas.drop(["Descuento", "Stock"], axis=1)
df_frutas
```

```python
df_frutas = df_frutas.assign(Descuento = [0.5, 0.4, 0.3, 0.3, 0.4, 0.5, 0.1, 0.2, 0.3],
                             Stock = [2, 2, 4, 5, 6, 1, 2, 3, 4])
df_frutas
```

```python
df_frutas = df_frutas.drop(["Descuento", "Stock"], axis=1)
df_frutas
```

```python
df_frutas.insert(2, "Stock", [2, 2, 4, 5, 6, 1, 2, 3, 4])
df_frutas
```

```python
df_frutas.insert(3, "Descuento", 0.5)
df_frutas
```

## actualizar celdas

```python
df_frutas.iloc[2,3] = 0.2
df_frutas
```

## renombrar columnas

```python
df_frutas.columns = ["Fruta", "Precio", "Existencias", "Descuento"]
df_frutas
```

## eliminar filas

```python
df_frutas = df_frutas.drop(7, axis=0)
df_frutas
```

## eliminar columnas

#### por nombre columna

```python
df_frutas = df_frutas.drop(["Precio"], axis=1)
df_frutas
```

#### por indice

```python
df_frutas = df_frutas.drop(df_frutas.columns[[1, 2]], axis = 1)
df_frutas
```

## limpiar (datos nulos)

```python
import numpy as np
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 1],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [2, 3, 4, 5],
                   [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))
df
```

#### reemplazar NaN

```python
df.fillna(0)
```

#### eliminar NaN

```python
# eliminar NaN por filas
df.dropna(axis=0)
```

```python
# eliminar NaN por columnas
df.dropna(axis=1)
```

### Convertir una matriz numpy a un dataframe de pandas

```python
import numpy as np
import pandas as pd
```

```python
matriz = np.array([[11,22,33,44],
                   [55,66,77,88]])
```

```python
matriz
```

```python
df = pd.DataFrame(matriz, columns=["a","b","c","d"])
```

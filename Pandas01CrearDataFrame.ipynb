### Crear DataFrame con lista o diccionario

```python
import pandas as pd
```

```python
lista_datos = [['edgar', 'hernandez', 26],
               ['moises', 'gonzalez', 25],
               ['juan', 'lopez', 18],
               ['francisco', 'perez', 4]
]
lista_columnas = ['nombre', 'apellido', 'edad']
```

```python
df = pd.DataFrame(lista_datos, columns=lista_columnas)
```

```python
import numpy as np
import pandas as pd
```

```python
# matriz 5x5 con numeros del -1 al 18 [-1,19)
matriz = np.random.randint(-1, 19, size=(5,5))
matriz
```

```python
df = pd.DataFrame(matriz, columns=["a","b","c","d","e"])
df
```

## minimo y maximo

```python
df.min()
```

```python
df["e"].min()
```

```python
df.min(axis=1)
```

```python
df.max()
```

```python
df["e"].max()
```

```python
df.max(axis=1)
```

## aplicar funciones

$ f(x) = 2x + 1 $

```python
def funcion(x):
    return 2 * x + 1
```

```python
# a todo el DataFrame
df = df.apply(funcion)
df
```

```python
# a una columna
df["b"] = df["b"].apply(funcion)
df
```

## funciones lambda

$ f(x) = x + 10 $

```python
y = lambda x: x + 10
print(y(5))

x = lambda a: a[0:3]
print(x("Funciones lambda en Python"))
```

$ f(x) = x^2 - 1 $

```python
# a una columna
df["b"] = df["b"].apply(lambda x: x * x - 1)
df
```

## Agrupar datos

```python
lista = [
         ["Cruz azul", "Mexico"],
         ["Tigres", "Mexico"],
         ["Barcelona", "España"],
         ["Real Madrid", "España"],
         ["Chelsea", "Inglaterra"],
         ["Manchester United", "Inglaterra"],
         ["Borussia Dortmund", "Alemania"],
         ["Juventus", "Italia"],
         ["Chivas", "Mexico"],
         ["Paris Saint-Germain", "Francia"],
         ["Sevilla", "España"],
         ["Olympique de Marsella", "Francia"],
         ["Boca Juniors", "Argentina"],
         ["Valencia", "España"],
         ["Liverpool", "Inglaterra"],
         ["Puebla", "Mexico"]
]
df = pd.DataFrame(lista, columns=["Equipo", "Pais"])
df
```

```python
grupo = df.groupby(["Pais"])
grupo.first()
```

```python
verdularia = [
         ["Manzana", "Fruta", 20],
         ["Papa", "Verdura", 20],
         ["Mango", "Fruta", 40],
         ["Melon", "Fruta", 10],
         ["Platano", "Fruta", 20],
         ["Naranja", "Fruta", 6],
         ["Zanahoria", "Verdura", 17],
         ["Brocoli", "Verdura", 30]
]
df = pd.DataFrame(verdularia, columns=["Alimento", "Categoria", "Precio"])
df
```

```python
grupo = df.groupby(["Categoria"])
grupo.first()
```

```python
grupo2 = df.groupby(["Categoria"]).mean()
grupo2
```
